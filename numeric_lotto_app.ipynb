```python
import numpy as np
import pandas as pd
```

```python
lst = np.random.randint(6,size = (6,6))
lst
sutun = ["1","2","3","4","5","6"]
lst
```

```python
satir = ["1.satir","2.satir","3.satir","4.satir","5.satir","6.satir",]
l_df = pd.DataFrame(lst,columns=(sutun),index = (satir))
l_df
```

```python
dz_1 = []
dz_2 = []
dz_3 = []
dz_4 = []
dz_5 = []
dz_6 = []

for i in range(0,6,1):
  dz_1.append(l_df.iloc[i][0])
dz_1

for i in range(0,6,1):
  dz_2.append(l_df.iloc[i][1])
dz_2

for i in range(0,6,1):
  dz_3.append(l_df.iloc[i][2])
dz_3

for i in range(0,6,1):
  dz_4.append(l_df.iloc[i][3])
dz_4

for i in range(0,6,1):
  dz_5.append(l_df.iloc[i][4])
dz_5

for i in range(0,6,1):
  dz_6.append(l_df.iloc[i][5])
dz_6
```

```python
series_dizi_1 = pd.Series(dz_1)
counts_1=series_dizi_1.value_counts()
counts_1

series_dizi_2 = pd.Series(dz_2)
counts_2=series_dizi_2.value_counts()
counts_2

series_dizi_3 = pd.Series(dz_3)
counts_3=series_dizi_3.value_counts()
counts_3

series_dizi_4 = pd.Series(dz_4)
counts_4=series_dizi_4.value_counts()
counts_4

series_dizi_5 = pd.Series(dz_5)
counts_5=series_dizi_5.value_counts()
counts_5

series_dizi_6 = pd.Series(dz_6)
counts_6=series_dizi_6.value_counts()
counts_6
```

```python
max_eleman_1=counts_1.idxmax()
max_eleman_2=counts_2.idxmax()
max_eleman_3=counts_3.idxmax()
max_eleman_4=counts_4.idxmax()
max_eleman_5=counts_5.idxmax()
max_eleman_6=counts_6.idxmax()
max_eleman_6
max_adet_1=counts_1[max_eleman_1]
max_adet_2=counts_2[max_eleman_2]
max_adet_3=counts_3[max_eleman_3]
max_adet_4=counts_4[max_eleman_4]
max_adet_5=counts_5[max_eleman_5]
max_adet_6=counts_6[max_eleman_6]
max_adet_6
```

```python
l_df
```

```python
tutan = 0
hedef = 3
if max_adet_1 >= hedef:
  tutan +=1
if max_adet_2 >= hedef:
  tutan +=1
if max_adet_3 >= hedef:
  tutan +=1
if max_adet_4 >= hedef:
  tutan +=1
if max_adet_5 >= hedef:
  tutan +=1
if max_adet_6 >= hedef:
  tutan +=1

print(f"tutan sayisi : {tutan}")
```
